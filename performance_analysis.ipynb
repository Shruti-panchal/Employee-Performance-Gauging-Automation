**Data Pipeline Connection**
Load Excel files
-

```python
# Importing all the necessary libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Data sheet checking and Exploring data (Are there any null values, or missing values?)
df = pd.read_excel("Dummy_Data.xlsx", sheet_name="Data")
# df.shape
# df.describe
# df.isnull().sum()
df
```

```python
# Loading Metrics Calculations sheet
met_df = pd.read_excel("Dummy_Data.xlsx", sheet_name="Metrics Calculations")
met_df
```

**Data Preprocessing**
Format Dates, 
Setting column names
-

```python
# Clearing the Mterics table (Column names fixing)
met_df.columns = met_df.iloc[0]
met_df = met_df[1:].reset_index(drop=True)
met_df['Target (Count)'] = pd.to_numeric(met_df['Target (Count)'])
met_df
```

```python
# Convert dates to datetime
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['End Date'] = pd.to_datetime(df['End Date'])
df.head()
```

**Metrics Calculations**
Performance Scoring
-

```python
# Grouping to get actual counts
employee_report_counts = (
    df.groupby(['Employee Code', 'Report Type']).size().reset_index(name='Actual Count'))

# Merge and calculate performance
fin_df = pd.merge(employee_report_counts, met_df, how='left',
                     left_on=['Employee Code', 'Report Type'],
                     right_on=['Employee Code', 'Metrics'])

fin_df['Performance %'] = round((fin_df['Actual Count'] / fin_df['Target (Count)']) * 100, 2)
```

**Final Scorecard**
-

```python
# Save DataFrame to Excel
file_name = "Final_Performance_Scorecard.xlsx"
fin_df.to_excel(file_name, index=False)

print(f"File saved successfully as {file_name}")
```

**Visualisation for Scorecard**
-

**Employee Performance Scorecard**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Employee Code", y="Performance %", hue="Report Type", data=fin_df)
plt.title("Employee Performance Scorecard")
plt.xlabel("Employee Code")
plt.ylabel("Performance %")
plt.show()
```

**Employee Performance Breakdow**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Employee Code", y="Actual Count", hue="Report Type", data=fin_df)
sns.lineplot(x="Employee Code", y="Target (Count)", data=fin_df, color="red", label="Target", linewidth=2)
plt.title("Employee Performance Breakdown")
plt.xlabel("Employee Code")
plt.ylabel("Reports Submitted")
plt.legend()
plt.show()
```

**Performance Score Distribution**

```python
plt.figure(figsize=(8, 5))
sns.histplot(fin_df["Performance %"], bins=10, kde=True, color="blue")
plt.title("Distribution of Performance Scores")
plt.xlabel("Performance Score (%)")
plt.ylabel("Frequency")
plt.show()
```

**Report Type Contribution per Employee**

```python
df_pie = fin_df.groupby(["Employee Code", "Report Type"])["Actual Count"].sum().reset_index()

for emp in df_pie["Employee Code"].unique():
    emp_data = df_pie[df_pie["Employee Code"] == emp]
    plt.figure(figsize=(6, 6))
    plt.pie(emp_data["Actual Count"], labels=emp_data["Report Type"], autopct="%1.1f%%", colors=["#ff9999","#66b3ff"])
    plt.title(f"Report Type Contribution - {emp}")
    plt.show()
```

**Trendline: Reports Generated Over Time**


```python
plt.figure(figsize=(10, 6))
df["Month"] = df["Start Date"].dt.strftime('%Y-%m')  # Extract Month-Year
monthly_counts = df.groupby(["Month", "Employee Code"]).size().reset_index(name="Report Count")
sns.lineplot(x="Month", y="Report Count", hue="Employee Code", data=monthly_counts, marker="o")
plt.xticks(rotation=45)
plt.title("Reports Generated Over Time")
plt.xlabel("Month")
plt.ylabel("Total Reports")
plt.show()
```
